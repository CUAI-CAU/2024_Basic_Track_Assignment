### 판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API
- **read_csv()**: CSV 파일 포맷 변환 API, 필드 구분 문자가 콤마(',')
  - CSV 뿐만 아니라 어떤 필드 구분 문자 기반의 파일 포맷도 변환 가능, 인자인 sep에 해당 구분 문자를 입력 (디폴트가 콤마)
  - 인자 filepath에 로드하려는 데이터 파일의 경로를 포함한 파일명을 입력
  - 별다른 파라미터 지정이 없으면 파일의 맨 처음 로우를 칼럼으로 변환
- **read_table()**: CSV 파일 포맷 변환 API, 필드 구분 문자가 탭('\t')
- **read_fwf()**: 고정 길이 기반의 칼럼 포맷을 DataFrame으로 로딩 API

### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
1) 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
   - **pd.DataFrame(data, columns)**
   - data: 리스트, 딕셔너리, 넘파이 ndarray
   - columns: 지정할 칼럼명 리스트 (지정하지 않으면 자동 할당)
   - DataFrame은 2차원 데이터이므로 2차원 이하의 데이터만 변환 가능

```python
import numpy as np

col_name1=['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)
# 리스트를 이용해 DataFrame 생성.
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)
# 넘파이 ndarray를 이용해 DataFrame 생성.
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
# 3개의 칼럼명이 필요함.
col_name2=['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape )
df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# Key는 컬럼명으로 매핑, Value는 리스트 형(또는 ndarray)
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

2) DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
   - **DataFrame.values**: DataFrame을 넘파이 ndarray로 변환
   - DataFrame.values.**tolist()**: DataFrame -> ndarray -> 리스트
   - DataFrame.values.**to_dict()**: DataFrame -> ndarray -> 딕셔너리
     - 인자로 'list'를 입력하면 딕셔너리의 값이 리스트형으로 반환됨

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
```

```python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### 정렬, Aggregation 함수, GroupBy 적용
- DataFrame, Series의 정렬 - **sort_values()**

  - ascending: True는 오름차순 (디폴트), False는 내림차순
  - inplace: False는 호출한 DataFrame은 그대로 유지하며 정렬된 DataFrame을 반환 (디폴트) / True는 호출한 DataFrame에 정렬 결과를 그대로 적용

- **groupby()** 적용

  - DataFrameGroupBy 타입의 객체를 반환
  - groupby()를 호출해 반환된 결과에 aggregation 함수를 호출하면 모든 칼럼에 해당 aggregation을 적용
  - groupby()**.agg()**: 적용하려는 여러 개의 aggregation 함수를 인자로 입력해서 사용, 딕셔너리 형태로 입력해 여러 개의 칼럼에 대해 서로 다른 aggregation 적용 가능

### apply lambda 식으로 데이터 가공

```python
# 함수 예제
def get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))
```

```python
# lambda 식 예제
lambda_square = lambda x : x ** 2
print('3의 제곱은:',lambda_square(3))
```

```python
# lambda 식 예제 - map() 함수 결합
a=[1,2,3]
squares = map(lambda x : x**2, a) # 리스트 a의 모든 요소를 제곱 계산
list(squares)
```

- DataFrame.**apply()** 에 lambda 식을 적용해 데이터를 가공
- lambda 식은 if else를 지원하는데, 이때 if 식보다 반환 값을 먼저 기술
- 세분화된 분류에는 함수를 사용하는 것이 나음

```python
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x)) # Name 칼럼의 문자열 개수를 나타내는 새로운 칼럼 'Name_len' 생성
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult') # 나이가 15 이하면 'Child', 그렇지 않으면 'Adult'를 나타내는 새로운 칼럼 'Child_Adult' 생성
titanic_df[['Age','Child_Adult']].head(8)
```

```python
# 나이가 15 이하면 'Child', 15 초과 60 이하면 'Adult', 61 이상이면 'Elderly'로 분류하는 새로운 칼럼 'Age_Cat' 생성
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 
                                                                                  'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성. 
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정. 
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
