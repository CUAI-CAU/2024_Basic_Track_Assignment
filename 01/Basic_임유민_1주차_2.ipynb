### 넘파이 ndarray, 리스트, 딕셔너리를 Dataframe으로 변환하기

```python
import numpy as np

col_name1 = ['col1']
#리스트 생성
list1= [1,2,3]

#ndarray 생성
array1 = np.array(list1)
print('array1 shape:', array1.shape)

#리스트를 이용해 Dataframe 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 Dataframe:\n', df_list1)

#넘파이 ndarray를 이용해 Dataframe 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 Dataframe:\n', df_array1)
```

### fillna()로 결곤 데이터 대체하기

```python
#Cabin 칼럼의 NaN 값을 C000 으로 대체
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
#Age 속성의 NaN 값을 Age 속성 값의 평균으로 대체
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())

#Embarked 속성의 NaN 값을 S로 대체
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')

#속성 별 결측치 합 반환
titanic_df.isna().sum()
```

# apply lambda 식으로 데이터 가공

```python
#input 값의 제곱 값을 반환하는 함수
ef get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))
```

```python
# lambda 를 이용하여 입력 값의 제곱을 반환하는 기능 구현
lambda_square = lambda x : x**2
print('3의 제곱은:',lambda_square(3))
```

```python
#lambda 식을 이용할 때 여러 개의 값을 입력 인자로 사용 => map() 함수 사용
a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
#Name 칼럼의 문자열 개수를 데이터 값으로 가지고 있는 Name_len 칼럼 형성
    #.apply() => DataFrame이나 Series에 일관된 연산을 적용할 때 사용하는 메서드 -> apply 뒤에 있는 연산을 앞의 dataframe, series 에 일괄 적용
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

```python
#Age 칼럼에서 15 미만이면 Child, 이상이면 Adult 인 Child_Adult 칼럼을 형성
    #주의 -> if 문보다 반환 값을 먼저 기술
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' )
titanic_df[['Age','Child_Adult']].head(8)
```

```python
#Age 칼럼에서 15 미만이면 Child, 15 이상 60 이하이면 Adult, 60 초과이면 Adult 인 Age_cat 칼럼을 형성
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 'Elderly'))

#Age_cat 칼럼의 값들의 갯수 추출
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
    # 조건에 따라 문자열로 cat return
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'

    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환 -> Age 칼럼 값에 따라 새로운 Age_cat 칼럼의 값을 get_category 함수의 반환 문자열 값으로 지정
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
