### 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

## apply lambda 식으로 데이터 가공

```python
def get_square(a): # 함수명과 입력 인자 선언
    return a ** 2 # 입력 인자 가공한 뒤 return 

print('3의 제곱은 :', get_square(3))
```

```python
lambda_square = lambda x : x ** 2 # ':' 로 입력 인자와 반환될 입력 인자의 계산식 분리
print('3의 제곱은 :', lambda_square(3)) 
```

```python
# 여러 개의 값을 입력 인자로 사용해야 할 경우에는 map() 함수

a = [1, 2, 3]
squares = map(lambda x : x ** 2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x)) # 'Name' 칼럼의 각 로우의 값을 입력인자로
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x < 15 else 'Adult') # if절보다 반환값을 먼저 기술
titanic_df[['Age', 'Child_Adult']]

# if else는 지원하지만 else if는 지원하지 않음
```

```python
# else if 를 이용하기 위해서는 else 절을 ()로 내포해 다시 if else 적용

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# else if 문이 많이 나오거나 switch case의 경우 별도의 함수를 만드는 게 더 나음
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.

def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'    
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
        
    return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(x)는 입력값으로 'Age' 칼럼 값 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
